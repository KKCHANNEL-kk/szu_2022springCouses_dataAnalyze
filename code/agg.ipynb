```python
import pandas as pd

raw = pd.read_csv('../data/allCourseScoreData.csv')
raw
```

```python
valid_date = ['2018-2019学年第一学期', '2018-2019学年第二学期', '2019-2020学年第一学期','2020-2021学年第一学期', '2020-2021学年第二学期']

data = raw[raw['semester'].isin(valid_date)]
data.dropna(axis=0,how='any',inplace=True)
data
```

```python
data['count_aplus'] = data.apply(lambda x:round(x['xkrs']*x['yxl']/100) , axis=1)
data
```

```python
data['avg_pjf'] = data.groupby(['KCH'])['pjf'].transform('mean')
data['avg_yxl'] = data.groupby(['KCH'])['yxl'].transform('mean')
data['avg_aplus'] = data.groupby(['KCH'])['count_aplus'].transform('mean')
data['max_aplus'] = data.groupby(['KCH'])['count_aplus'].transform('max')
data['max_yxl'] = data.groupby(['KCH'])['yxl'].transform('max')
data['min_yxl'] = data.groupby(['KCH'])['yxl'].transform('min')
data['avg_xkrs'] = data.groupby(['KCH'])['xkrs'].transform('mean')
data['avg_jgl'] = data.groupby(['KCH'])['jgl'].transform('mean')
data['min_jgl'] = data.groupby(['KCH'])['jgl'].transform('min')
data
```

```python
result = data.drop_duplicates(['KCH']).drop(['semester','xkrs','jgl','pjf','yxl','count_aplus'],axis=1)
result
```

```python
result.to_csv('../data/aggScoreData.csv',index=False,encoding='utf-8-sig')
```
