```python
import pandas as pd

base = pd.read_csv('../data/SpringCourseBasic.csv')
base
```

```python
score = pd.read_csv('../data/aggScoreData.csv')
```

```python
raw_result = pd.merge(base, score,how='left', left_on='courseNumber',right_on='KCH')
```

```python
strdata1 = raw_result.groupby('courseNumber',as_index=False)['teacherName'].agg(lambda x:x.str.cat(sep=',')).reset_index()
strdata2 = raw_result.groupby('courseNumber',as_index=False)['teachingPlace'].agg(lambda x:x.str.cat(sep='\n')).reset_index()
```

```python
raw_result= pd.merge(raw_result, strdata1,how='left', on='courseNumber')
raw_result= pd.merge(raw_result, strdata2,how='left', on='courseNumber')
```

```python
dropdup_result = raw_result.drop_duplicates(subset=['courseNumber']).drop(['KCH','teacherName_x','teachingClassID','teachingPlace_x'],axis=1)
```

```python
dropdup_result.rename(columns={'teacherName_y':'teacherName','teachingPlace_y':'teachingPlace'},inplace=True)
```

```python
dropdup_result = dropdup_result.drop(['index_x','index_y'],axis=1).reset_index(drop=True)
```

```python
dropdup_result.to_csv('../data/SpringCourseBasic_with_score.csv',index=False,encoding='utf-8-sig')
```
